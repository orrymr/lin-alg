```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
cov = np.array([[1, 0], 
                [0, 1]])

mean = np.array([0, 0])

data = np.random.multivariate_normal(mean, cov, 500)

print(cov)
print("*****")
print(mean)
```

```python
plt.scatter(data[:, 0], data[:, 1], s = 2)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
```

<h2> Let's apply a transformation to this data </h2>

```python
transformation1 = np.array([[4, 2], 
                            [2, 5]])

data_transformed = data.dot(transformation1)

plt.scatter(data_transformed[:, 0], data_transformed[:, 1], s = 2)
plt.xlim(-20, 20)
plt.ylim(-20, 20)
```

```python
transformation2 = np.array([[4, -2], 
                            [-2, 5]])

data_transformed = data.dot(transformation2)

plt.scatter(data_transformed[:, 0], data_transformed[:, 1], s = 2)
plt.xlim(-20, 20)
plt.ylim(-20, 20)
```

```python
transformation3 = np.array([[4, 5], 
                            [5, 5]])

data_transformed = data.dot(transformation3)

plt.scatter(data_transformed[:, 0], data_transformed[:, 1], s = 2)
plt.xlim(-20, 20)
plt.ylim(-20, 20)
```

```python
# 0 Covariance between the components in this example

transformation4 = np.array([[4, 0], 
                            [0, 1]])

data_transformed = data.dot(transformation4)

plt.scatter(data_transformed[:, 0], data_transformed[:, 1], s = 2)
plt.xlim(-20, 20)
plt.ylim(-20, 20)
```

```python
# 0 Covariance between the components in this example as well

transformation5 = np.array([[1, 0], 
                            [0, 4]])

data_transformed = data.dot(transformation5)

plt.scatter(data_transformed[:, 0], data_transformed[:, 1], s = 2)
plt.xlim(-20, 20)
plt.ylim(-20, 20)
```

<h2> Let's see how the various covariance matrices affect a vector </h2>

```python
V = np.array([[0,1],[1,1],[1,0]])
origin = [0], [0] # origin point


fig = plt.figure()

plt.subplot(2, 2, 1)
plt.quiver(*origin, V[:,0], V[:,1], color=['r','b','g'], scale=5)


plt.subplot(2, 2, 2)
V_ = V.dot(np.array([[2, 0], 
                    [0, 2]]))
plt.quiver(*origin, V_[:,0], V_[:,1], color=['r','b','g'], scale=5)


plt.subplot(2, 2, 3)
V_ = V.dot(np.array([[1, 2], 
                    [2, 1]]))
plt.quiver(*origin, V_[:,0], V_[:,1], color=['r','b','g'], scale=5)


plt.subplot(2, 2, 4)
V_ = V.dot(np.array([[1, -2], 
                    [-2, 1]]))
plt.quiver(*origin, V_[:,0], V_[:,1], color=['r','b','g'], scale=5)


```

<h2> OK, let's do some Eigenvalue stuff... </h2>

But before we do, let's think a little bit about the covariance matrix.
It is a symmetric matrix. Since it is a symmetric matrix, its eigenvalues are real and its eigenvectors are orthogonal.



Let's look at the case where the covariance matrix of the data has 0's along the diagonal.

```python
cov = np.array([[4, 0], 
                [0, 3]])

mean = np.array([0, 0])

data = np.random.multivariate_normal(mean, cov, 500)

plt.scatter(data[:, 0], data[:, 1], s = 2)
plt.xlim(-20, 20)
plt.ylim(-20, 20)
```

So, if we get the eigenvalues and eigenvectors of the covariance matrix, then the eigenvalues should just be equal to the variances in the x and y directions

```python
values, vectors = np.linalg.eig(cov)

print(values)
print(vectors)
```

Look how the eigenvectors above are orthogonal. This is because the covariance matrix is symmetric.

Let's plot the e-vectors on top of the data:

```python
plt.scatter(data[:, 0], data[:, 1], s = 2)
origin = [0], [0] # origin point
plt.quiver(*origin, vectors[:,0], vectors[:,1], color=['r','b'])
plt.xlim(-20, 20)
plt.ylim(-20, 20)
```

Let's use data with non zeros along the diagonal. This means that there is variance in direction other than x and y axes.

```python
cov = np.array([[4, 2], 
                [2, 3]])

mean = np.array([0, 0])

data = np.random.multivariate_normal(mean, cov, 500)

plt.scatter(data[:, 0], data[:, 1], s = 2)
plt.xlim(-20, 20)
plt.ylim(-20, 20)
```

Let us get the eigenvalues / eigenvectors of the covariance matrix...

```python
values, vectors = np.linalg.eig(cov)

print(values)
print(vectors)
```

```python
vectors[:,0].dot(vectors[:,1]) # e-vectors are still orthogonal... covariance matrix is always symmetric.
```
